Instalando pacotes necessários

```python
import numpy as np # pacote para cálculo numérico
import pandas as pd # pandas para puxar em conjunto dados em tabela
import matplotlib.pyplot as plt # função so para fazer grafico
import seaborn as sns # função mais fácil do matplitlib
```

Seaborn já vem com conjunto de dados

```python
tips=sns.load_dataset('tips')
```

Função para ver os dados

```python
tips.head()
```

Diagrama de dispersão (gorgeta e o valor da conta são proporcionais)

```python
sns.scatterplot(x='total_bill',y='tip', data=tips)
```

```python
sns.scatterplot(x='total_bill',y='tip', data=tips,hue='day') # hue muda a cor dos dias da semana
```

```python
sns.scatterplot(x='total_bill',y='tip', data=tips,size='size') # aumenta as bolinhas conforma o valor da gorjeta (size)
```

```python
sns.scatterplot(x='total_bill',y='tip', data=tips,hue='day',palette='magma') # palette da cores diferentes de uma paleta so com hue
```

Boxplot

```python
sns.boxplot(x='day',y='tip',data=tips)
```

```python
sns.boxplot(x='day',y='tip',hue='sex',data=tips) # gorjeta por dia da semana
```

Carregar outro conjunto de dados

Mapa de Calor (correlação)

```python
sns.barplot(x="sex",y="total_bill",data=tips) # media de quem paga mais caro na conta
```
